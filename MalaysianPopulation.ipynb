```python
population['pop_5'] + population['pop_12'] + population['pop_60'] + population['adult'] == population['pop']
```

```python
population['babies'] = population['pop'] - (population['pop_5'] + population['pop_12'] + population['pop_60'] + population['adult'])
population
```

```python
population['babies'] + population['pop_5'] + population['pop_12'] + population['pop_60'] + population['adult'] == population['pop']
```

```python
population = population[['state','babies','pop_5','pop_12','adult','pop_60', 'pop']]
population
```

```python
q1 = failedbanklist.describe().loc['25%', 'Fund']
q1
```

```python
q3 = failedbanklist.describe().loc['75%', 'Fund']
```

```python
q3
```

```python
iqr = q3 - q1
iqr
```

```python
iqrmin = q1 - (1.5 * iqr)
iqrmin
```

```python
iqrmax = q3 + (1.5 * iqr)
iqrmax
```

```python
failedbanklist[(failedbanklist['Fund'] < iqrmin)]
```
